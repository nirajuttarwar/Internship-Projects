## **Univariate Analysis**

### **Histograms of all data**

### **Histogram for Numeric Data**

```python
numeric_col = [col for col in train.columns if train[col].dtype != 'object']

# Calculate the number of rows needed for subplots
num_rows = (len(numeric_col) + 1) // 2  # +1 to handle odd number of numeric vars

plt.figure(figsize=(15, 4 * num_rows))  # Adjust figure height based on num_rows

for i, var in enumerate(numeric_col):
    plt.subplot(num_rows, 2, i + 1)  # 2 columns per row
    sns.histplot(train[var], bins=30, kde=False)  # Plot histogram without KDE
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Histogram is mainly used for the analyzing the skewness of the data. In the Data the histogram is showing the skewness of data.

### **Count Plot for Categorical Data**

```python
categorical_col = [col for col in train.columns if train[col].dtype == 'object']

plt.figure(figsize=(15, 12))
for i, col in enumerate(categorical_col, 1):
  if col != 'NObeyesdad':
    plt.subplot(3, 3, i)
    sns.countplot(data=train, x=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```

There are no balance between categories in the same feature.

The majority of people consume large quantities of food that are high in calories (FAVC).

Most individuals consume food between meals (CAEC) sometimes.

We observe a significant difference in the prevalence of smoking among individuals, with varying proportions of smokers and non-smokers. The same pattern is noted for the SCC feature.

Most individuals consume alcoholic beverages (CALC) sometimes.

As the vast majority of individuals opt for public transportation as their primary mode of commuting, we will delve deeper into the MTRANS feature by examining the distribution of weight classifications.

```python
class_distribution = train['NObeyesdad'].value_counts(normalize=True)

class_distribution.plot(kind='bar', color='skyblue')
plt.xlabel('Class')
plt.ylabel('Proportion')
plt.title('Class Distribution in Target Variable')
plt.xticks(rotation=45)
plt.show()
```

There are 7 classes in total, representing different obesity levels or weight categories. The classes seem to be relatively evenly distributed, with proportions ranging approximately from 0.10 to 0.20.

```python
for col in categorical_col:
  if col != 'NObeyesdad':
    train[col].value_counts().reset_index(name="Total")
    fig=px.pie(train[col].value_counts().reset_index(name="Total"),values="Total",names=col,hole=.6,template="plotly_dark",title=col)

    fig.update_traces(hoverinfo='label+percent', textinfo='label+percent', textfont_size=20)
    fig.show()
```

## **Bivariate Analysis**

### **Scatter Plot**

```python
plt.figure(figsize=(15, 5))
sns.scatterplot(y='Age', x='Weight', hue='NObeyesdad', data=train)
plt.title('Scatter Plot Example')
plt.ylabel('Age')
plt.xlabel('Weight')
plt.show()
```

As in the above plot shows that 20 to 30 age groups individuals fall into higher obesity categories and also individuals having insufficient weight falls into 14 to 25 age group.

So it conclude that 14 to 30 age groups peoples need to pay attentation to their health.

```python
plt.figure(figsize=(15, 5))
sns.scatterplot(x='Height', y='Weight', hue='NObeyesdad', data=train)
plt.title('Scatter Plot Example')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

Normal Weight and Insufficient Weight: These categories (green and red dots) tend to be clustered around lower weights (40-70 kg) and heights ranging from 1.5 to 1.75 meters.

Overweight Levels: Overweight Level I & II (orange and light blue) tend to have a higher spread in both height and weight, generally between 70-120 kg and heights between 1.5 to 1.9 meters.

Obesity Types I to III: Obesity Type I, II, and III (purple, pink, and dark blue) appear at much higher weights, from 80 kg to over 140 kg, with varying heights. These individuals with higher obesity levels tend to have weights well above 100 kg.

Height Distribution: People with higher weights tend to have a wider height distribution, though most individuals are between 1.5 to 1.9 meters tall.

```python
plt.figure(figsize=(40, 6))
plt.subplot(1, 3, 1)
sns.countplot(x='NObeyesdad', hue='family_history_with_overweight', data=train)
plt.title('Obesity Levels by Family History')
```

People with a family history of being overweight (shown in blue) are much more likely to fall into severe obesity categories, particularly Obesity Type III and Obesity Type II. These groups have the highest counts among individuals with a family history.
A noticeable portion of individuals without a family history (shown in orange) fall into the Normal Weight and Insufficient Weight categories. However, there are still many people with a family history of obesity who manage to maintain normal weight levels.
In both Overweight Level I and Overweight Level II, the majority of individuals have a family history of being overweight. However, there is still a small percentage without a family history who are overweight.

```python
plt.figure(figsize=(40, 6))
plt.subplot(1, 3, 2)
sns.boxplot(x='NObeyesdad', y='FAF', data=train)
plt.title('Exercise Frequency (FAF) by Obesity Level')
```

People in the Obesity Type III and Obesity Type II categories have the lowest exercise frequency. The median for these groups is below 1 hour per week, and the overall spread of exercise time is smaller compared to other groups.

This suggests that a lack of physical activity may be a significant contributor to severe obesity levels.

The Normal Weight and Insufficient Weight categories have the highest median exercise frequency, around 1.5 to 2 hours per week. These groups also have wider ranges of activity, indicating more variability in exercise habits.

People classified as Overweight Level I and Overweight Level II also have moderate exercise habits, with medians around 1 hour per week. Their exercise patterns are more similar to the normal weight category than the obesity groups.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=train, x='FAVC', hue='NObeyesdad', palette='Set2')

# Adding labels and title
plt.title('Count Plot of High-Calorie Food Consumption (FAVC) and Obesity Risk (NObeyesdad)', fontsize=16)
plt.xlabel('High-Calorie Food Consumption (FAVC)', fontsize=14)
plt.ylabel('Count of Individuals', fontsize=14)
plt.legend(title='Obesity Categories', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.tight_layout()
plt.show()
```

The majority of individuals across all obesity categories consume high-calorie food.

There is a clear association between high-calorie food consumption and higher obesity levels, as shown by the larger counts in the obese categories for individuals who consume high-calorie foods.

The plot suggests that interventions to reduce high-calorie food intake may have the most significant impact on individuals in higher obesity categories.

```python
grouped_df = train.groupby("Gender")["NObeyesdad"].value_counts().reset_index(name="Count")

fig = px.bar(grouped_df,
             x="Gender",
             y="Count",
             color="NObeyesdad",
             barmode="group",
             title="Count of Obesity Levels by Gender")
fig.show()
```

Severe obesity (Obesity_Type_III) is significantly more common among females, while moderate obesity (Obesity_Type_I and II) is more common among males.

Being underweight (Insufficient_Weight) is more prevalent in females.
The Normal_Weight category is almost evenly distributed between both genders.
